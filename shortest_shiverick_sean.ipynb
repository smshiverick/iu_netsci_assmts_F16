### 1. Write Shortest Path Algorithm
Write algorithm that calculates the shortest paths between all the nodes 
then place the lengths of those shortest paths in a sequence (such as a list or array).

#### First: Create a function to calculate shortest path lengths from single node to everyone else:
(1) You need a queue that stores the nodes to visit at the next hop. 
(2) You pull one from this queue, say i, assign the current distance, and 
(3) then go through each neighbor of i and see whether we have seen the node before or not. 
(4) If we have seen it, we can ignore it (we already know the shortest distance). 
(5) If not, we should put this node to the queue. 
Note: You may want to have two queues (or lists or sets) for the "current" level and the "next" level of the search.

```python
# Breadth First Search algorithm by D. Eppstein (2007; see also LexBFS.py)
# http://www.ics.uci.edu/~eppstein/PADS/BFS.py 

start = {}
def all_shortest_paths(random_graph, start):
    visited = set()
    currentLevel = [start]
    while currentLevel:
        for v in currentLevel:
            visited.add(v)
        nextLevel = set()
        levelGraph = {v:set() for v in currentLevel}
        for v in currentLevel:
            for w in random_graph[v]:
                if w not in visited:
                    levelGraph[v].add(w)
                    nextLevel.add(w)
        yield levelGraph
        currentLevel = nextLevel
```
