```python
# Create Random graph based on Les Miserables character network

import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline

random_graph = nx.erdos_renyi_graph(77, 0.09)
print(nx.info(random_graph))
```

```python
# Draw the graph and display plot; write the plot as .gml file

nx.draw(random_graph)
plt.show

nx.write_gml(random_graph, "random_lesmis.gml")
```
